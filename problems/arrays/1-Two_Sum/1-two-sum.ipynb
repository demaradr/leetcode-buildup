## Many ways to solve

| Approach | Idea | Time |
|----------|------|------|
| **Brute force** | Try every pair `(i, j)` and check if they sum to `target` | O(n²) |
| **Hash map** (what you'll build) | As you scan, remember each number's index; check if the **complement** was seen already | O(n) |

Brute force is easy to write but slow on large inputs. **The hash map approach is the best time complexity** — one pass through the array, O(n) time.

## Why a hash map?

You need two numbers that add up to `target`. For the current number `n`, the other number you need is:

```python
diff = target - n   # the "complement"
```

If you've **already seen** `diff` earlier in the array, you found your pair.

A dictionary lets you answer *"have I seen this value before, and at what index?"* in O(1) time:

```python
prevMap = {}   # value : index

prevMap[7] = 1   # saw 7 at index 1
1 in prevMap     # False — we store values as keys, not indices
7 in prevMap     # True
prevMap[7]       # → 1
```

As you loop:
1. Compute `diff = target - n`
2. If `diff in prevMap` → return `[prevMap[diff], i]` (the old index and current index)
3. Otherwise → `prevMap[n] = i` and keep going

You store **after** checking so you don't use the same element twice.
