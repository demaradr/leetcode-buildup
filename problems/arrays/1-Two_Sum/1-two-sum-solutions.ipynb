## Many ways to solve

| Approach | Idea | Time |
|----------|------|------|
| **Brute force** | Check every pair of indices | O(n²) |
| **Hash map** (below) | Remember each value's index; look up `target - n` | O(n) |

**The hash map approach is the best time complexity** — one pass, O(n) time, O(n) space for the map.

## Why `prevMap`?

For each number `n`, the complement is `diff = target - n`. If `diff` is already a key in `prevMap`, we found the pair: `[prevMap[diff], i]`.

We map **value → index** because the problem asks for indices. We check **before** storing `n` so we never pair an element with itself.

```python
nums = [2, 7, 11, 15]
target = 9
prevMap = {}

for i, n in enumerate(nums):
    diff = target - n
    if diff in prevMap:
        print(f"found pair: [{prevMap[diff]}, {i}]")
        break
    prevMap[n] = i

```

### Exercise 11 — Wrap it in a function


```python
def two_sum(nums, target):
    prevMap = {}

    for i, n in enumerate(nums):
        diff = target - n
        if diff in prevMap:
            return [prevMap[diff], i]
        prevMap[n] = i


print(two_sum([2, 7, 11, 15], 9))
print(two_sum([3, 3], 6))

```

## Part 2 — Full solution

O(n) hash map — store each value's index, look up the complement `target - n` before adding the current number.


```python
from typing import List


class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        prevMap = {}  # val : index

        for i, n in enumerate(nums):
            diff = target - n
            if diff in prevMap:
                return [prevMap[diff], i]
            prevMap[n] = i
        return

```

```python
sol = Solution()

assert sol.twoSum([2, 7, 11, 15], 9) == [0, 1]
assert sol.twoSum([3, 2, 4], 6) == [1, 2]
assert sol.twoSum([3, 3], 6) == [0, 1]

print("All tests passed.")

```
