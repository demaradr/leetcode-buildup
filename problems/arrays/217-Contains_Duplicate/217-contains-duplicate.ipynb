## Many ways to solve

There is more than one way to solve this problem:

| Approach | Idea | Time |
|----------|------|------|
| **Brute force** | Compare every pair of numbers | O(n²) |
| **Sorting** | Sort first; duplicates end up next to each other | O(n log n) |
| **Hash set** (what you'll build) | Remember each number as you go; check if you've seen it before | O(n) |

Brute force works on tiny inputs, but with up to 10⁵ numbers it is too slow. **The hash set approach is the best time complexity** — one pass through the array, O(n) time.

## Why a set?

A **set** only stores **unique** values — adding the same number twice does not create a duplicate entry. That matches exactly what we care about: *has this number already appeared?*

As you loop through `nums`:
1. If `num` is **already in** the set → we've seen it before → duplicate → return `True`
2. Otherwise → add `num` to the set and keep going

Checking `num in hashset` is fast (average O(1)), so each number costs constant time. That's why we use a set instead of a list of seen numbers, where `in` would scan the whole list each time.
