## Many ways to solve

| Approach | Idea | Time |
|----------|------|------|
| **Brute force** | Compare every pair of numbers | O(n²) |
| **Sorting** | Sort first; duplicates end up next to each other | O(n log n) |
| **Hash set** (below) | Remember each number as you go; check if you've seen it before | O(n) |

Brute force works on tiny inputs, but with up to 10⁵ numbers it is too slow. **The hash set approach is the best time complexity** — one pass, O(n) time.

## Why a set?

A **set** only stores **unique** values — adding the same number twice does not create a duplicate entry. That matches what we need: *has this number already appeared?*

1. If `num` is **already in** the set → duplicate → return `True`
2. Otherwise → add `num` and continue

`num in hashset` is O(1) on average, so each element is handled in constant time. A list of seen numbers would require scanning the whole list on every check.

### Exercise 6 — Spot a duplicate while looping


```python
nums = [1, 2, 3, 1]
seen = set()

for num in nums:
    if num in seen:
        print(f"duplicate: {num}")
    else:
        seen.add(num)

```

### Exercise 7 — Stop early on a duplicate


```python
nums = [1, 2, 3, 1]
seen = set()

for num in nums:
    if num in seen:
        print("found duplicate")
        break
    seen.add(num)

```

### Exercise 8 — Return a boolean


```python
nums = [1, 2, 3, 1]
seen = set()
found = False

for num in nums:
    if num in seen:
        found = True
        break
    seen.add(num)

print(found)

nums = [1, 2, 3, 4]
seen = set()
found = False

for num in nums:
    if num in seen:
        found = True
        break
    seen.add(num)

print(found)

```

### Exercise 9 — Wrap it in a function


```python
def has_duplicate(nums):
    seen = set()
    for num in nums:
        if num in seen:
            return True
        seen.add(num)
    return False


print(has_duplicate([1, 2, 3, 1]))
print(has_duplicate([1, 2, 3, 4]))

```

## Part 2 — Full solution

This is the O(n) hash set solution — the best time complexity for this problem. We use a set because it rejects duplicates automatically and gives fast membership checks (`num in hashset`), so we can detect a repeat in one pass.


```python
from typing import List


class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        hashset = set()
        for num in nums:
            if num in hashset:
                return True
            hashset.add(num)
        return False

```

```python
sol = Solution()

assert sol.containsDuplicate([1, 2, 3, 1]) is True
assert sol.containsDuplicate([1, 2, 3, 4]) is False
assert sol.containsDuplicate([1, 1, 1, 3, 3, 4, 3, 2, 4, 2]) is True
assert sol.containsDuplicate([1]) is False

print("All tests passed.")

```
