## Many ways to solve

| Approach | Idea | Time |
|----------|------|------|
| **Brute force** | For each char in `s`, try to find and remove a match in `t` | O(n²) |
| **Sorting** | Sort both strings; anagrams become identical | O(n log n) |
| **Hash map** (what you'll build) | Count how many times each letter appears in each string; compare counts | O(n) |

Sorting is simple and fine for short strings. **The hash map approach is the best time complexity** — two passes to build counts, one pass to compare, O(n) time.

## Why a dictionary (hash map)?

An anagram uses the **same letters with the same frequencies** — order does not matter. A **dictionary** maps each character to **how many times** it appears:

```python
"aab" → {"a": 2, "b": 1}
```

Unlike a set (which only tracks *presence*), a dict tracks **counts**. Two strings are anagrams if and only if their count dictionaries match.

The pattern you'll use over and over:

```python
count[c] = 1 + count.get(c, 0)
```

- `count.get(c, 0)` — "how many `c` have I seen so far? Default to 0 if never seen."
- Add 1 and store it back under `c`.
