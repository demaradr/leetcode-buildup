## Many ways to solve

| Approach | Idea | Time |
|----------|------|------|
| **Brute force** | Match and remove chars one by one | O(n²) |
| **Sorting** | Sort both strings and compare | O(n log n) |
| **Hash map** (below) | Count letter frequencies; compare dicts | O(n) |

**The hash map approach is the best time complexity** — O(n) time, O(1) space for the 26 lowercase letters.

## Why a dictionary?

Anagrams have identical **character counts**. A dict maps each letter → frequency. The `.get(c, 0)` pattern safely increments counts even when the key is new.

```python
word = "aabb"
counts = {}

for char in word:
    counts[char] = 1 + counts.get(char, 0)

print(counts)

```

### Exercise 12 — Put it in a function


```python
def is_anagram(s, t):
    if len(s) != len(t):
        return False

    countS, countT = {}, {}

    for i in range(len(s)):
        countS[s[i]] = 1 + countS.get(s[i], 0)
        countT[t[i]] = 1 + countT.get(t[i], 0)

    for c in countS:
        if countS[c] != countT.get(c, 0):
            return False
    return True


print(is_anagram("anagram", "nagaram"))
print(is_anagram("aacc", "ccac"))

```

## Part 2 — Full solution

O(n) hash map solution — count letter frequencies in both strings, then compare.


```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False

        countS, countT = {}, {}

        for i in range(len(s)):
            countS[s[i]] = 1 + countS.get(s[i], 0)
            countT[t[i]] = 1 + countT.get(t[i], 0)

        for c in countS:
            if countS[c] != countT.get(c, 0):
                return False
        return True

```

```python
sol = Solution()

assert sol.isAnagram("anagram", "nagaram") is True
assert sol.isAnagram("rat", "car") is False
assert sol.isAnagram("aacc", "ccac") is False
assert sol.isAnagram("a", "a") is True
assert sol.isAnagram("ab", "a") is False

print("All tests passed.")

```
